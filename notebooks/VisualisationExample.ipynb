```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drinks = ["Cappuccino", "HotChocolate", "Tea", "Black Coffee", "Mocha", "Espresso"]
sales =  [91, 76, 27, 66, 52, 56]
tea_df = pd.DataFrame(sales, index=drinks, columns=['Total Sales'])
```

```python
# Days of the week:
days = [1, 2, 3, 4, 5, 6,7]
# Money spend one week 1
money_spent = [1000, 1200, 1500, 1080, 1400, 1650, 1350]
# Money spend one week 2
money_spent_2 = [900, 1500, 1200, 1050, 950, 1250, 1000]
# Create figure:
fig = plt.figure(figsize=(10,5))
# Plot first week expenses:
plt.plot(days, money_spent)
# Plot second week expenses:
plt.plot(days, money_spent_2)
# Display the result:
plt.title('Company Expenditure Comparison')
plt.legend(['First week', 'Second week'])
plt.show()
```

```python
# Sales and regions:
region = ['US', 'UK','SA','CA','AU']
sales = [7000000,430000,670000,150000,27000]
colors = ['blue', 'steelblue', 'gray','red','green']
explode = (0.1, 0.2, 0, 0, 0.0) 
# Create figure and plot pie:
fig = plt.figure(figsize=(10,8))
plt.pie(sales, labels=region,autopct='%d%%', colors=colors,explode=explode)
plt.axis('equal')
plt.title('Corona Cases Worldwide', fontsize='20')
#plt.savefig('plot_eight.png')
plt.show()
```
