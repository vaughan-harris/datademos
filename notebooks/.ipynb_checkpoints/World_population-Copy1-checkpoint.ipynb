```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
## open the CSV file
datapath="/home/vaughan/Documents/Data/github/population/data/population.csv"
population = pd.read_csv(datapath)
countries = sorted(population['Country Name'].unique())
countries=['United Kingdom','United States']
```

```python
x=1
country_data = population[(population['Country Name'] == countries[x])]

pop=country_data['Value']
grw=country_data['Growth']
date=country_data['Year']
diff=pop.diff()
#type(country_data)
#country_data.head()
```

```python
Chart_Title = "Population in " + countries[x]

fig = plt.figure(figsize=(12,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title(Chart_Title)
ax.grid(False)
ax.yaxis.grid(True)



ax.plot(date, pop, linewidth="2", marker='o')



#ax = plt.subplot(111)

#ax.bar(x, z, width=0.5, color='g', align='center')
#ax.bar(x, k, width=0.5, color='r', align='center')
#ax.xaxis_date()

ax.set_ylabel('population numbers', 
                     color='blue', 
                     size='x-large') 
ax_growth  = ax.twinx()

ax_growth.set_ylabel('population growth', 
                     color='blue', 
                     size='x-large') 

ax_growth.tick_params(axis='y', 
                      labelcolor='blue',
                      labelsize='large')

ax_growth.bar(date, diff, width=0.5, color='c', alpha=0.5, align='center')

```

# Multiple plots on a figure

```python

fig = plt.figure(figsize=(16,16)) 
x=0
country_data = population[(population['Country Name'] == countries[x])]
pop=country_data['Value']
ukpop=country_data['Value']
grw=country_data['Growth']
date=country_data['Year']
diff=pop.diff()

ax1 = fig.add_subplot(211)
ax1.plot(date, pop, linewidth="2", marker='o')
Chart_Title = "Population in " + countries[x]
ax1.set_title(Chart_Title)
ax1_growth  = ax1.twinx()
ax1_growth.set_ylabel('population growth', 
                     color='blue', 
                     size='x-large') 
ax1_growth.tick_params(axis='y', 
                      labelcolor='blue',
                      labelsize='large')
ax1_growth.bar(date, diff, width=0.5, color='c', alpha=0.5, align='center')





## ===============================================================================

x=1
country_data = population[(population['Country Name'] == countries[x])]
pop=country_data['Value']
uspop=country_data['Value']
grw=country_data['Growth']
date=country_data['Year']
diff=pop.diff()

ax2 = fig.add_subplot(212)
ax2.plot(date, pop, linewidth="2", marker='o', color="orange")
Chart_Title = "Population in " + countries[x]
ax2.set_title(Chart_Title)

ax2_growth  = ax2.twinx()
ax2_growth.set_ylabel('population growth', color='blue', size='x-large') 
ax2_growth.tick_params(axis='y', labelcolor='blue',labelsize='large')
ax2_growth.bar(date, diff, width=0.5, color='c', alpha=0.5, align='center')
```

```python
#https://stackoverflow.com/questions/38897091/axes-subplot-y-size

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.2)
y = np.sin(x)

f, (a0, a1) = plt.subplots(2,1, gridspec_kw = {'height_ratios':[1, 4]}, sharex=True)  # <---- sharex=True will share the xaxis between the two axes
a0.plot(x, y)
a1.plot(y, x)
plt.show()
```
